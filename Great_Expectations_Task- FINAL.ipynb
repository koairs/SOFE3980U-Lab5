##5. Set Up Great Expectations Context and Data Source

```python
context = gx.get_context()

data_source = context.data_sources.add_pandas(name="labels_datasource")
data_asset = data_source.add_dataframe_asset(name="labels_asset")
```

##6. Define and Create a Data Batch

```python
batch_definition = data_asset.add_batch_definition_whole_dataframe(
    name="labels_batch_definition"
)

batch = batch_definition.get_batch(batch_parameters={"dataframe": df})
```

### Expectation 1

```python
expectation_1 = gx.expectations.ExpectColumnValuesToBeUnique(
    column="Timestamp"
)
```

### Validate Data Against Expectation 1

```python
validation_result_1 = batch.validate(expectation_1)
validation_result_1
```

### Expectation 2

```python
expectation_2 = gx.expectations.ExpectColumnValuesToBeBetween(
    column="pedestrianLocationX_TopLeft",
    min_value=0,
    max_value=1920
)
```
