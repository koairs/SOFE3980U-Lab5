##3. Load Labels.csv

Download and upload the [Labels.csv](https://github.com/zubxxr/SOFE3980U-Lab5/blob/main/Labels.csv) into this notebook, and then load the file.

```python
df.head()
```

##5. Set Up Great Expectations Context and Data Source

```python
context = gx.get_context()
data_source = context.sources.add_pandas(name="labels_pandas_datasource")
data_asset = data_source.add_dataframe_asset(name="labels_asset")
```

##6. Define and Create a Data Batch

```python
batch_request = data_asset.build_batch_request(dataframe=df)
```

##7. Define Three Expectations for Column Values

Using this [link](https://greatexpectations.io/expectations/), choose three expectation functions and apply them to the labels dataset in a relevant manner.

You should replace the 'ExpectColumnValuesToBeBetween' function with other functions you select from the link.

You can also check the format/parameters required of each function when you click "See more" on the function.

```python
## Original Function
expectation = gx.expectations.ExpectColumnValuesToBeBetween(
    column="column", min_value=0, max_value=20
)

## Example Function

## This function only requires a column parameter, and not a max or min value
expectation = gx.expectations.ExpectColumnValuesToBeUnique(
    column="column"
)
```

### Expectation 1

```python
expectation_unique = gx.expectations.ExpectColumnValuesToBeUnique(column="Timestamp")
```

### Validate Data Against Expectation 1

```python
validation_result_1 = context.definitions.get_validator(
    batch_request=batch_request, 
    expectation_suite_name="unique_timestamps"
).validate(expectation_unique)

print(f"Expectation 1 Success: {validation_result_1.success}")
```

### Expectation 2

```python
expectation_range = gx.expectations.ExpectColumnValuesToBeBetween(
    column="pedestrianLocationX_TopLeft", min_value=0, max_value=1000
)
```

### Validate Data Against Expectation 2

```python
validation_result_2 = context.definitions.get_validator(
    batch_request=batch_request, 
    expectation_suite_name="coordinate_range"
).validate(expectation_range)

print(f"Expectation 2 Success: {validation_result_2.success}")
```

### Expectation 3

```python
expectation_not_null = gx.expectations.ExpectColumnValuesToNotBeNull(column="Ground_Truth_View")
```

### Validate Data Against Expectation 3

```python
validation_result_3 = context.definitions.get_validator(
    batch_request=batch_request, 
    expectation_suite_name="non_null_views"
).validate(expectation_not_null)

print(f"Expectation 3 Success: {validation_result_3.success}")
```
